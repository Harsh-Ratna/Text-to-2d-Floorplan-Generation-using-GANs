```python
import os
import pandas as pd

# Define the directories containing the .txt and .png files
text_directory = "human_annotated_tags"
image_directory = "../floorplan_image"

# Initialize lists to store file paths, texts, and image paths
file_names = []
texts = []
image_paths = []

# Iterate through each text file in the directory
for filename in os.listdir(text_directory):
    if filename.endswith(".txt"):
        # Extract the file name without extension
        file_name = os.path.splitext(filename)[0]
        # Construct the file paths
        text_file_path = os.path.join(text_directory, filename)
        image_file_path = os.path.join(image_directory, f"{file_name}.png")
        # Read the text from the file
        with open(text_file_path, "r", encoding="utf-8") as file:
            text = file.read()
        # Append the file name, text, and image path to the respective lists
        file_names.append(file_name)
        texts.append(text)
        image_paths.append(image_file_path)

# Create a DataFrame from the lists
df = pd.DataFrame({"File_Name": file_names, "Text": texts, "Image_Path": image_paths})
```

```python
# df contains 4000 image paths of floor plans and its text descriptions.
import pandas as pd
from transformers import BertTokenizer, BertModel
import torch

# Load pre-trained BERT model and tokenizer
tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')
model = BertModel.from_pretrained('bert-base-uncased')

# Function to tokenize text and get BERT embeddings
def get_bert_embeddings(text):
    # Tokenize input text
    inputs = tokenizer(text, return_tensors='pt', padding=True, truncation=True)
    
    # Forward pass through BERT model
    with torch.no_grad():
        outputs = model(**inputs)
    
    # Extract hidden states (last layer) of BERT
    hidden_states = outputs.last_hidden_state
    
    # Average pooling of hidden states to get sentence embeddings
    sentence_embeddings = torch.mean(hidden_states, dim=1)
    
    return sentence_embeddings.numpy()


# Apply BERT embeddings extraction function to the 'Text' column
df['BERT_Embeddings'] = df['Text'].apply(get_bert_embeddings)

# Display DataFrame with BERT embeddings
print(df)
```
